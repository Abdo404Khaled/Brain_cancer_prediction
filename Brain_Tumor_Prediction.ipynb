```python
import matplotlib.pyplot as plt
from matplotlib.image import imread

import cv2

import random

from sklearn.model_selection import train_test_split

from os import listdir
import numpy as np
```

```python
folder = 'train/'
for i in range(9):
    plt.subplot(330 + 1 + i)
    filename = folder + 'y' + str(i) + '.jpg'
    image = cv2.imread(filename, cv2.IMREAD_COLOR)
    plt.imshow(image)
plt.show()
```

```python
folder = 'train/'
for i in range(9):
    plt.subplot(330 + 1 + i)
    filename = folder + 'no' + str(i) + '.jpg'
    image = cv2.imread(filename, cv2.IMREAD_COLOR)
    plt.imshow(image)
plt.show()
```

```python
folder = 'train/'
photos= []
for file in listdir(folder):
    label = 0
    if file.startswith('y'):
         label = 1
    try:
        photo = cv2.imread(folder + file, cv2.IMREAD_GRAYSCALE)
        new_photo = cv2.resize(photo, (200,200))
        photos.append([new_photo, label])
    except Exception as e:
        pass
```

```python
random.shuffle(photos)
```

```python
X = []
y = []

for feature, label in photos:
    X.append(feature)
    y.append(label)

X = np.array(X).reshape(-1, 200, 200, 1)
y = np.array(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2)
```

```python
from keras.layers import Dense, Conv2D, MaxPooling2D, Flatten
from keras import Sequential


model = Sequential()

model.add(Conv2D(256, kernel_size=(3, 3), activation='relu', input_shape=(200, 200, 1)))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(256, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())

model.add(Dense(64, activation='relu'))

model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)
```

```python
model.summary()
```

```python
model.fit(
    X_train,
    y_train,
    epochs = 3,
    batch_size=16
)
```

```python
model.evaluate(X_test, y_test)
```

```python
model.save('Brain_Tumor.h5')
```
